```python
#Tutorial # 03

## The central limit theorem holds that the sum of a large number of identically 
## distributed random numbers will be normally distributed no matter how the 
## individual numbers are distributed.
##
## In lecture we proved this and saw how the pdf for the sum of variables 
## can be found numerically from the pdf of the individuals using the Fourier 
## transform.
##
## Now we will verify the central limit theorem by numerically experiment :

# 1) Create 1000 uniform random numbers between -1 and 1 using numpy.rand

       # make histogram of numbers
import numpy as np
import matplotlib.pyplot as plt
import math
from scipy.stats import norm
x=np.random.rand(1000)
z=2*x-1
plt.hist(z,normed=True,bins=20)
plt.title('Distribution of 1000 uniform random numbers')
plt.show()
# 2) Make 1000 sums of 3 uniform random numbers between -1 and 1

       # make histogram of numbers

       # draw normal distribution with the same man and variance over histogram 
       # (What should the variance be?)
s=[]
for j in range(1000):
    i=0
    y=[]
    while i<3:
        i=i+1
        y.append(2*np.random.rand() - 1)
       # print(y)
    s.append(np.sum(y))
# print(s)
var=np.var(s)
mean=np.mean(s)
sigma = math.sqrt(var)
x1 = np.linspace(mean-3*sigma,mean+3*sigma, 100)
plt.plot(x1,norm.pdf(x1, loc=mean,scale=sigma),color='red',label='Sums of 3 numbers')
plt.hist(s,bins=20,normed=True,color='red',alpha=0.4)
# 3) Make 1000 sums of 10 uniform random numbers between -1 and 1

       # make histogram of numbers

       # draw normal distribution with the same variance over histogram
s1=[]
for j1 in range(1000):
    i1=0
    y1=[]
    while i1<10:
        i1=i1+1
        y1.append(2*np.random.rand() - 1)
       # print(y)
    s1.append(np.sum(y1))
# print(s)
var1=np.var(s1)
mean1=np.mean(s1)
sigma1 = math.sqrt(var1)
x2 = np.linspace(mean1-3*sigma1,mean1+3*sigma1, 100)
plt.title('Distribution of 1000 sums of 3 and 10 uniform random numbers')
plt.plot(x2,norm.pdf(x2, loc=mean1,scale=sigma1),color='blue',label='Sums of 10 numbers')
plt.hist(s1,bins=20,normed=True,color='blue',alpha=0.4)
plt.legend(loc='upper right',fontsize='small')
plt.show()
# 4) Do 1 through 3 for a lognormal distribution instead of uniform.  The 
#    lognormal distribution can be found in the scipy.stats package. 
r=np.random.lognormal(size=1000)
plt.hist(r,normed=True,bins=20)
plt.title('Distribution of 1000 lognormally distributed random numbers')
plt.show()

b=[]
for a in range(1000):
    r = np.random.lognormal(size=3)
    #print(r)
    b.append(np.sum(r))
    #print(np.sum(r))
#print(b)
var2=np.var(b)
mean2=np.mean(b)
sigma2=math.sqrt(var2)
x3 = np.linspace(mean2-3*sigma2,mean2+3*sigma2, 100)
plt.plot(x3,norm.pdf(x3, loc=mean2,scale=sigma2),color='red',label='Sums of 3 numbers')
plt.hist(b,bins=20,normed=True,color='red',alpha=0.4)

b1=[]
for a1 in range(1000):
    r1 = np.random.lognormal(size=10)
    #print(r)
    b1.append(np.sum(r1))
    #print(np.sum(r))
#print(b)
var3=np.var(b1)
mean3=np.mean(b1)
sigma3=math.sqrt(var3)
x4 = np.linspace(mean3-3*sigma3,mean3+3*sigma3, 100)
plt.title('Distribution of 1000 sums of 3 and 10 lognormally distributed random numbers')
plt.plot(x4,norm.pdf(x4, loc=mean3,scale=sigma3),color='blue',label='Sums of 10 numbers')
plt.hist(b1,bins=20,normed=True,color='blue',alpha=0.4)
plt.legend(loc='upper right')
plt.show()

```
