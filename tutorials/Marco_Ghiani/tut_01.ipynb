```python
#Tutorial number 1:-
#-- Create a scatter plot of a gaussian data (hint: see numpy funtion random.normal) of size 1000,
#-- located at 0.0 and with scale 5.0.
#-- Enhance the plot with axis labels  (x-position , y-position) and don't forget the units in Mpc/h, 
#-- plot title ("2d galaxy distribution") and x&y-axis ticks [-5, -3, .., 3, 5] and set dimond as point marker, 
#-- blue as point color and  1.5 for  point size. 

#-- Create a histogram of the previous x&y data with 20 equispaced linear bins and plot it. 
#-- Creat a combined plot that shows the scatter data in the middel with x&y historgrams beside each axis. 
#-- Look at the follwoing example for help (https://matplotlib.org/examples/pylab_examples/scatter_hist.html)
import numpy as np
import matplotlib.pyplot as plt
plt.figure(num=None,figsize=(4,4),dpi=300)
plt.xlabel("x-position [Mpc/h]" )
plt.ylabel("y-position [Mpc/h]")
plt.title ("2d galaxy distribution")
x_tick = [-17,-15,-13,-11,-9,-7, -5 , -3 , -1 , 1 , 3, 5 , 7, 9, 11, 13, 15, 17 ]
y_tick = [-17,-15,-13,-11,-9,-7, -5 , -3 , -1 , 1 , 3, 5 , 7, 9, 11, 13, 15, 17 ]
plt.xticks(x_tick)
plt.yticks(y_tick) 
functionx = np.random.normal(loc=0.0, scale= 5.0, size=1000 )
functiony = np.random.normal(loc=0.0, scale= 5.0, size=1000 )
plt.scatter( functionx,functiony , c= 'blue', marker= 'd' , s=1.5)
plt.show()




```

```python
marco= np.linspace(min(functionx),max(functionx),20)
marcoy= np.linspace(min(functiony),max(functiony),20)
plt.figure(num=None,figsize=(3,3),dpi=300)
plt.hist(functionx, bins=marco, edgecolor='blue',histtype='step',label='x' )
plt.hist(functiony, bins=marcoy, edgecolor='red' ,histtype='step', label='y')
plt.legend()
plt.show()
```

```python
import matplotlib.gridspec as gridspec
grid=gridspec.GridSpec(3,3)
plt.figure(num=None,figsize=(4,4),dpi=300)
plt.subplots_adjust(hspace=0.0,wspace=0.0)
x1=plt.subplot(grid[0:1,0:2])
x2=plt.subplot(grid[1:3,0:2])
x3=plt.subplot(grid[1:, 2:3])
x2.scatter(functionx,functiony , c= 'violet', marker= 'd' , s=1.5)
x1.hist(functionx, bins=marco, edgecolor='blue',histtype='step' )
x3.hist(functiony, bins=marcoy, edgecolor='red' ,histtype='step', orientation='horizontal')
x2.set_xlabel("x-position [Mpc/h]")
x2.set_ylabel("y-position [Mpc/h]")
x1.set_title("2d galaxy distribution")
x1.tick_params(labelbottom='off')
x3.tick_params(labelleft='off')

plt.show()

               
            
```
