```python
#Tutorial # 03

## The central limit theorem holds that the sum of a large number of identically 
## distributed random numbers will be normally distributed no matter how the 
## individual numbers are distributed.
##
## In lecture we proved this and saw how the pdf for the sum of variables 
## can be found numerically from the pdf of the individuals using the Fourier 
## transform.
##
## Now we will verify the central limit theorem by numerically experiment :

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as ss

%matplotlib inline

fig, axs = plt.subplots(1, 3, figsize=(15,5))

#$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$
# 1) Create 1000 uniform random numbers between -1 and 1 using numpy.rand
rand_unif_num = np.random.uniform(-1.0, 1.0, 1000)

# make histogram of numbers
plt.subplot(1, 3, 1)
plt.hist(rand_unif_num, alpha=0.7, normed=True, edgecolor='yellow',)
plt.title('Uniformly random numbers')

#$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$
# 2) Make 1000 sums of 3 uniform random numbers between -1 and 1
a = np.random.uniform(-1.0, 1.0, 1000)
b = np.random.uniform(-1.0, 1.0, 1000)
c = np.random.uniform(-1.0, 1.0, 1000)
sums_3 = a + b + c

# make histogram of numbers
plt.subplot(1, 3, 2)
n, bins, patches = plt.hist(sums_3, alpha=0.7, normed=True, color='yellow', edgecolor='blue', label='Sums')

# draw normal distribution with the same man and variance over histogram 
# (What should the variance be?)
mean = sums_3.mean()
sigma = np.sqrt(sums_3.var())
x_values = np.linspace(np.amin(bins), np.amax(bins), 100)
plt.plot(x_values, ss.norm.pdf(x_values, loc=mean, scale=sigma), color='blue', label='Normal')

plt.legend()
plt.title('Sums of 3 numbers')

#$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$
# 3) Make 1000 sums of 10 uniform random numbers between -1 and 1
mega = []
sums_10 = np.zeros(1000)
for i in range(10):
    mega.append(np.random.uniform(-1.0, 1.0, 1000))    
    mega_np=np.array(mega[i])
    sums_10 += mega_np

# make histogram of numbers
plt.subplot(1, 3, 3)
n, bins, patches = plt.hist(sums_10, normed=True, alpha=0.7, color='yellow', edgecolor='blue', label='Sums')

# draw normal distribution with the same variance over histogram
mean = sums_10.mean()
sigma = np.sqrt(sums_10.var())
x_values = np.linspace(np.amin(bins), np.amax(bins), 100)
plt.plot(x_values, ss.norm.pdf(x_values, loc=mean, scale=sigma), color='blue', label='Normal')

plt.legend()
plt.title('Sums of 10 numbers')

plt.tight_layout()
plt.show()

#$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$
# 4) Do 1 through 3 for a lognormal distribution instead of uniform.  The 
#    lognormal distribution can be found in the scipy.stats package. 

fig, axs = plt.subplots(1, 3, figsize=(15,5))

# 1) Create 1000 lognormal random numbers with mu=0 and sigma=1 using numpy.rand
rand_log_num = np.random.lognormal(0.0, 1.0, 1000)

# make histogram of numbers
plt.subplot(1, 3, 1)
plt.hist(rand_log_num, bins=30, alpha=0.7, normed=True, edgecolor='yellow',)
plt.title('Lognormal random numbers')

# 2) Make 1000 sums of 3 lognormal random numbers with mu=0 and sigma=1
a = np.random.lognormal(0.0, 1.0, 1000)
b = np.random.lognormal(0.0, 1.0, 1000)
c = np.random.lognormal(0.0, 1.0, 1000)
sums_3 = a + b + c

# make histogram of numbers
plt.subplot(1, 3, 2)
n, bins, patches = plt.hist(sums_3, bins=30, alpha=0.7, normed=True, color='yellow', edgecolor='blue', label='Sums')

# draw normal distribution with the same man and variance over histogram 
# (What should the variance be?)
mean = sums_3.mean()
sigma = np.sqrt(sums_3.var())
x_values = np.linspace(np.amin(bins), np.amax(bins), 100)
plt.plot(x_values, ss.norm.pdf(x_values, loc=mean, scale=sigma), color='blue', label='Normal')

plt.legend()
plt.title('Sums of 3 numbers')

# 3) Make 1000 sums of 10 lognormal random numbers with mu=0 and sigma=1
mega = []
sums_10 = np.zeros(1000)
for i in range(10):
    mega.append(np.random.lognormal(0.0, 1.0, 1000))    
    mega_np=np.array(mega[i])
    sums_10 += mega_np

# make histogram of numbers
plt.subplot(1, 3, 3)
n, bins, patches = plt.hist(sums_10, bins=30, normed=True, alpha=0.7, color='yellow', edgecolor='blue', label='Sums')

# draw normal distribution with the same variance over histogram
mean = sums_10.mean()
sigma = np.sqrt(sums_10.var())
x_values = np.linspace(np.amin(bins), np.amax(bins), 100)
plt.plot(x_values, ss.norm.pdf(x_values, loc=mean, scale=sigma), color='blue', label='Normal')

plt.legend()
plt.title('Sums of 10 numbers')

plt.tight_layout()
plt.show()
```
