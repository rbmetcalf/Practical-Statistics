```python
# homework # 01

## Here we are going to get a little experience with the covariance matrix 
## of data and principle components analysis (PCA).  This was introduced in the 
## lectures.

# 1) Read file homework_01_2d-datafile.csv into a dataframe using pandas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data_2d = pd.read_csv('homework_01_2d_datafile.csv')

print(data_2d.head())
    # make a scatter plot of them
_ = plt.scatter(data_2d['X'], data_2d['Y'])
_ = plt.xlabel('X')
_ = plt.ylabel('Y')
plt.show()

# 2) Find the covariance matrix for the two variables.  There is a python function 
#    for this, but it would be more instructive if you write your own.

x_mean = data_2d['X'].mean()
x_std = data_2d['X'].std()
y_mean = data_2d['Y'].mean()
y_std = data_2d['Y'].std()
s = 0

for i in range(len(data_2d['X'])):
    s = s + ((data_2d['X'][i]-x_mean)*(data_2d['Y'][i]-y_mean))

cov = s/(len(data_2d['X'])*x_std*y_std)
    # Are they correlated? (Remember to normalize the correlation coefficient.)
    
print(cov)

# 3) Decompose the covariance matrix using an eigenvalue decompositions (There are 
#    python functions for this.)

    # What are the principle components of the data? Normalize the vectors to one.
x = ( data_2d['X'] - x_mean)/ x_std
y = ( data_2d['Y'] - y_mean)/ y_std
    # What are the variances of each principle component?
print('the covariance matrix is:')
print(np.cov(x,y))
# 4) Do 1 through 3, but using the data file homework_01_5d-datafile.csv this time.
#    In this case the data is 5 dimensional.

data_5d = pd.read_csv('homework_01_5d_datafile.csv')

print(data_5d.head())
    # make a scatter plot of them
_ = plt.scatter(data_5d['X1'], data_5d['X2'])
_ = plt.xlabel('X1')
_ = plt.ylabel('X2')
plt.show()

_ = plt.scatter(data_5d['X1'], data_5d['X3'])
_ = plt.xlabel('X1')
_ = plt.ylabel('X3')
plt.show()

_ = plt.scatter(data_5d['X1'], data_5d['X4'])
_ = plt.xlabel('X1')
_ = plt.ylabel('X4')
plt.show()

_ = plt.scatter(data_5d['X1'], data_5d['X5'])
_ = plt.xlabel('X1')
_ = plt.ylabel('X5')
plt.show()

_ = plt.scatter(data_5d['X2'], data_5d['X3'])
_ = plt.xlabel('X2')
_ = plt.ylabel('X3')
plt.show()

_ = plt.scatter(data_5d['X2'], data_5d['X4'])
_ = plt.xlabel('X2')
_ = plt.ylabel('X4')
plt.show()

_ = plt.scatter(data_5d['X2'], data_5d['X5'])
_ = plt.xlabel('X2')
_ = plt.ylabel('X5')
plt.show()

_ = plt.scatter(data_5d['X3'], data_5d['X4'])
_ = plt.xlabel('X3')
_ = plt.ylabel('X4')
plt.show()

_ = plt.scatter(data_5d['X3'], data_5d['X5'])
_ = plt.xlabel('X3')
_ = plt.ylabel('X5')
plt.show()

_ = plt.scatter(data_5d['X4'], data_5d['X5'])
_ = plt.xlabel('X4')
_ = plt.ylabel('X5')
plt.show()



x1 = ( data_5d['X1'] - data_5d['X1'].mean())/ data_5d['X1'].std()
x2 = ( data_5d['X2'] - data_5d['X2'].mean())/ data_5d['X2'].std()
x3 = ( data_5d['X3'] - data_5d['X3'].mean())/ data_5d['X3'].std()
x4 = ( data_5d['X4'] - data_5d['X4'].mean())/ data_5d['X4'].std()
x5 = ( data_5d['X5'] - data_5d['X5'].mean())/ data_5d['X5'].std()


    # What are the variances of each principle component?
print('the covariance matrix between x1 and x2 is:')
print(np.cov(x1,x2))

print('the covariance matrix between x1 and x3 is:')
print(np.cov(x1,x3))

print('the covariance matrix between x1 and x4 is:')
print(np.cov(x1,x4))

print('the covariance matrix between x1 and x5 is:')
print(np.cov(x1,x5))

print('the covariance matrix between x2 and x3 is:')
print(np.cov(x2,x3))

print('the covariance matrix between x2 and x4 is:')
print(np.cov(x2,x4))

print('the covariance matrix between x2 and x5 is:')
print(np.cov(x2,x5))

print('the covariance matrix between x3 and x4 is:')
print(np.cov(x3,x4))

print('the covariance matrix between x3 and x5 is:')
print(np.cov(x3,x5))

print('the covariance matrix between x4 and x5 is:')
print(np.cov(x4,x5))

    
```
