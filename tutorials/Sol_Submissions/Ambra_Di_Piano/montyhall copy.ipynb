```python
#########################################
#### FIRST TRIAL IN INTERACTIVE MODE ####
#########################################

from numpy.random import randint
import numpy as np

n_doors = 3
available_doors = [0,1,2]
remain_doors = available_doors


# set prized door
win_door = np.random.choice(available_doors)

# pick a door
print('Now we will go together through the first tiral! Take a pick!! Enter a number between 0 and 2')
choice = input()
choice = int(choice)
print('So you choose the door numer ', choice, '!')

# delate chosen door from list of available
rm_door = available_doors.index(choice)
rm_door = int(rm_door)

del remain_doors [rm_door]

# set changing pick loop for not opening the winning or the chosen door
Loop = True

while Loop == True :
    
    # set the goat behind a door
    goat_door = np.random.choice(remain_doors) 
    if goat_door != win_door :

        print('Now I will open a door for you, behind which there is no prize but a very sad goat.. ')
        print('Said door is the number: ', goat_door)
        Loop = False

    else :
        print('*roll of drums*')

# change chance
print('Would you bet your chances changing you pick? Enter yes or no!')
change = input()


if change == 'yes' :
    print('Choose another door please, enter a number between 0 and 2')
    print('Remember the goat is behind door numer ', goat_door, ' and your first choice was number ', choice)
    choice_2 = input()
    choice_2 = int(choice_2)
    
    # giving result
    if choice_2 == choice :
        print('So you do not want to change you pick after all..')
    elif choice_2 == goat_door :
        print('Here was the goat though, sorry you lost..')
    elif choice_2 == win_door :
        print('You won!!')
    else :
        print('Lets see..')
    
elif change == 'no' :
    print('You stick with your pick, lets see what there is behind the door!')
    
    if choice == win_door :
        print('You won!!')

    else :
        print('Sorry.. you lost. Here is another goat.. ')

 
    
# giving results
else :
    print('Please enter just "yes" or "no"!')
    change = input()
   
#################################
#### LOOPING THROUGH TRIALAS ####
#################################

print('Now lets go iterative for few Monty Hall picks and see what chances come out of it!')
print('How many trials you wanna go through?')
trials_choice = input()
trials_choice = int(trials_choice)

results = np.zeros(2)

for i in range(0,trials_choice) :
    winner = randint(0,3)
    choice = randint(0,3)
    
    if(choice == winner) :
        # don't switch
        results[0] = results[0] +1
        
    else :
        # do switch
        results[1] = results[1] +1
        
print('Chances of winning if you do not change your pick and if you do (respectively):  ', results/np.sum(results))
```
