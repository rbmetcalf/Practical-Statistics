```python
# homework # 01

## Here we are going to get a little experience with the covariance matrix 
## of data and principle components analysis (PCA).  This was introduced in the 
## lectures.

# 1) Read file homework_01_2d-datafile.csv into a dataframe using pandas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import csv

data2D = pd.read_csv('homework_01_2d-datafile.csv', ',', header=0)
#print(data2D)

    # make a scatter plot of them

data2D = data2D.iloc[:,1:]
#print(data2D)
    
x = data2D.loc[:,'X']
y = data2D.loc[:,'Y']

#print(x, '----', y)
    
_ = plt.scatter(x, y, marker='D', c='red', alpha=0.5)
_ = plt.title('2D data scatter plot')
_ = plt.xlabel('x')
_ = plt.ylabel('y')

plt.show()

# 2) Find the covariance matrix for the two variables.  There is a python function 
#    for this, but it would be more instructive if you write your own.  
    
cov2D = np.cov(data2D, rowvar=False)
cov2D_norm = np.corrcoef(data2D, rowvar=False)
print('covariance matrix: \n', cov2D, '\n'*2, 'normed correlation matrix: \n', cov2D_norm)
      
    # Are they correlated? (Remember to normalize the correlation coefficient.)
    
print('\n The observations are correlated.')

# 3) Decompose the covariance matrix using an eigenvalue decompositions (There are 
#    python functions for this.)

eig_vals2D, eig_vecs2D = np.linalg.eig(cov2D_norm)
print('\n'*2, 'eigenvalues: \n', eig_vals2D, '\n'*2, 'eigenvectores: \n', eig_vecs2D)

```

```python
    # What are the principle components of the data? Normalize the vectors to one.
eig_vecs2D_norm = eig_vecs2D / (np.linalg.norm(eig_vecs2D)/2)
print('\n'*2, 'eigenvectors: \n', eig_vecs2D, '\n'*2, 'normed eigenvectores: \n', eig_vecs2D_norm)

   
    # What are the variances of each principle component?
    
# list of eig_vals e eig_vecs touple
eig_pairs2D = [(eig_vals2D[i], eig_vecs2D[i]) for i in range(len(eig_vals2D))]
eig_pairs2D.sort()
eig_pairs2D.reverse()

print('\ncheck sorting of eigenvalues:')

for i in eig_pairs2D : print(i[0])
    

```
