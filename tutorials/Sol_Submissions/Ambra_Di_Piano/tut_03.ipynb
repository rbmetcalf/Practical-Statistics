```python
#Tutorial # 03

## The central limit theorem holds that the sum of a large number of identically 
## distributed random numbers will be normally distributed no matter how the 
## individual numbers are distributed.
##
## In lecture we proved this and saw how the pdf for the sum of variables 
## can be found numerically from the pdf of the individuals using the Fourier 
## transform.
##
## Now we will verify the central limit theorem by numerically experiment :

# 1) Create 1000 uniform random numbers between -1 and 1 using numpy.rand
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm
from scipy.stats import lognorm

dist = np.empty(1000)
sum_dist3 = np.empty(1000)
sum_dist10 = np.empty(1000)
log_dist3 = np.empty(1000)
log_dist10 = np.empty(1000)

for i in range(1000) :
    dist[i] = np.random.rand() 
#dist = np.random.random_sample(1000) in range(1)
#print(dist[0:50])

sns.set() 
# make histogram of the numbers
_ = plt.hist(dist, normed=True)
_ = plt.title('random number sample in [0,1)')
_ = plt.xlabel('normed random number')
_ = plt.ylabel('normed counts')

plt.show()

# 2) Make 1000 sums of 3 uniform random numbers between -1 and 1
for i in range(1000) :
    sum_dist3[i] = sum(np.random.rand() for j in range(3))
    
#print(sum_dist)       
# make histogram of the numbers
    
_ = plt.hist(sum_dist3, normed=True)
_ = plt.title('sample of sum of 3 random namber in [0,1)')
_ = plt.xlabel('sum of 3 random number')
_ = plt.ylabel('normed counts')

plt.show()

# draw the pdf for a normal distribution with the same man and variance
# over histogram (What should the variance be?)
x = np.arange(sum_dist3.min(), sum_dist3.max(), 0.001 )
_ = plt.plot(x, norm.pdf(x, sum_dist3.mean(), sum_dist3.std()), label='norm.pdf')
_ = plt.hist(sum_dist3, normed=True, label='random collection')
_ = plt.title('sample of sum of 3 random namber in [0,1)')
_ = plt.xlabel('sum of 3 random numer')
_ = plt.ylabel('normed counts')
_ = plt.legend(loc=0)

plt.show()

# 3) Make 1000 sums of 10 uniform random numbers between -1 and 1
for i in range(1000) :
    sum_dist10[i] = sum(np.random.rand() for j in range(10))
    
# make histogram of the numbers
_ = plt.hist(sum_dist10, normed=True)
_ = plt.title('sample of sum of 10 random namber in [0,1)')
_ = plt.xlabel('sum of 10 random number')
_ = plt.ylabel('normed counts')

plt.show()

# check
#x2 = np.arange(sum_dist10.min(), sum_dist10.max(), 0.001 )
#_ = plt.plot(x2, norm.pdf(x2, sum_dist10.mean(), sum_dist10.std()), label='norm.pdf')
#plt.show()

# draw the pdf for a normal distribution with the same variance over histogram
x2 = np.arange(sum_dist10.min(), sum_dist10.max(), 0.001 )
_ = plt.plot(x2, norm.pdf(x2, sum_dist10.mean(), sum_dist10.std()), label='norm.pdf')
_ = plt.hist(sum_dist10, normed=True, label='random collection')
_ = plt.title('sample of sum of 10 random namber in [0,1)')
_ = plt.xlabel('sum of 10 random numer')
_ = plt.ylabel('normed counts')
_ = plt.legend(loc=0)

plt.show()

# 4) Do 1) through 3) for a lognormal distribution instead of uniform.  The 
#    lognormal distribution can be found in the scipy.stats package. Use 
#    the case s = 1.  And normalize the sum with 1/N where N is the number of 
#    numbers in each sum.  Draw the Gaussian with the same mean and variance as 
#    the sample.

for i in range(1000) :
    log_dist3[i] = sum(np.random.lognormal(size=1) for j in range(3)) 
    log_dist10[i] = sum(np.random.lognormal(size=1) for j in range(10))
    
log_dist3 = log_dist3 / 1000
log_dist10 = log_dist10 / 1000

_ = plt.hist(log_dist3, normed=True)
_ = plt.title('sample of sum of 3 random namber (normed 1/3) in [0,1)')
_ = plt.xlabel('sum of 3 random number')
_ = plt.ylabel('normed counts')

plt.show()

x3 = np.arange(log_dist3.min(), log_dist3.max(), 0.001 )
_ = plt.plot(x3, norm.pdf(x3, log_dist3.mean(), log_dist3.std()), label='lognorm.pdf')
_ = plt.hist(log_dist3, normed=True, label='random collection')
_ = plt.title('sample of sum of 3 random namber (normed 1/3) in [0,1)')
_ = plt.xlabel('sum of 3 random numer')
_ = plt.ylabel('normed counts')
_ = plt.legend(loc=0)

plt.show()

_ = plt.hist(log_dist10, normed=True)
_ = plt.title('sample of sum of 10 random namber (normed 1/10) in [0,1)')
_ = plt.xlabel('sum of 10 random number')
_ = plt.ylabel('normed counts')

plt.show()

x4 = np.arange(log_dist10.min(), log_dist10.max(), 0.001 )
_ = plt.plot(x4, norm.pdf(x4, log_dist10.mean(), log_dist10.std()), label='norm.pdf')
_ = plt.hist(log_dist10, normed=True, label='lograndom collection')
_ = plt.title('sample of sum of 10 random namber (normed 1/10) in [0,1)')
_ = plt.xlabel('sum of 10 random numer')
_ = plt.ylabel('normed counts')
_ = plt.legend(loc=0)

plt.show()
```
