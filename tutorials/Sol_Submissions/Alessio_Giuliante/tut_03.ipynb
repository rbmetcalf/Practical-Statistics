```python
#Tutorial # 03

## The central limit theorem holds that the sum of a large number of identically 
## distributed random numbers will be normally distributed no matter how the 
## individual numbers are distributed.
##
## In lecture we proved this and saw how the pdf for the sum of variables 
## can be found numerically from the pdf of the individuals using the Fourier 
## transform.
##
## Now we will verify the central limit theorem by numerically experiment :

# 1) Create 1000 uniform random numbers between -1 and 1 using numpy.rand

       # make histogram of numbers
    
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

x=np.random.uniform(-1, 1, 1000)
plt.hist(x)
plt.show()


# 2) Make 1000 sums of 3 uniform random numbers between -1 and 1

       # make histogram of numbers

       # draw normal distribution with the same man and variance over histogram 
       # (What should the variance be?)
    
s=[]   
for i in range(1000): 
    s.append(np.random.uniform(-1,1)+np.random.uniform(-1,1)+np.random.uniform(-1,1))
       
plt.hist(s, normed=True)

m=np.mean(s)
v=np.var(s)
y = np.linspace(m-3*v, m+3*v, 100)
plt.plot(y, norm.pdf(y, loc=m, scale=np.sqrt(v)), c='red', lw=2)

plt.show()


# 3) Make 1000 sums of 10 uniform random numbers between -1 and 1

       # make histogram of numbers

       # draw normal distribution with the same variance over histogram
        
t=[]
for j in range(1000): 
    t.append(np.random.uniform(-1,1)+np.random.uniform(-1,1)+np.random.uniform(-1,1)+np.random.uniform(-1,1)+np.random.uniform(-1,1)+np.random.uniform(-1,1)+np.random.uniform(-1,1)+np.random.uniform(-1,1)+np.random.uniform(-1,1)+np.random.uniform(-1,1))
       
plt.hist(t, normed=True)

n=np.mean(t)
z=np.var(t)
w = np.linspace(n-2*z, n+2*z, 100)
plt.plot(w, norm.pdf(w, loc=n, scale=np.sqrt(z)), c='red', lw=2)

plt.show()


# 4) Do 1 through 3 for a lognormal distribution instead of uniform.  The 
#    lognormal distribution can be found in the scipy.stats package. 

g=np.random.lognormal(-1, 1, 1000)
plt.hist(g)
plt.show()
    
b=[]   
for k in range(1000): 
    b.append(np.random.lognormal(-1,1)+np.random.lognormal(-1,1)+np.random.lognormal(-1,1))
       
plt.hist(b, normed=True)

p=np.mean(b)
q=np.var(b)
u = np.linspace(p-2*q, p+5*q, 100)
plt.plot(u, norm.pdf(u, loc=p, scale=np.sqrt(q)), c='red', lw=2)

plt.show()
        
c=[]
for l in range(1000): 
    c.append(np.random.lognormal(-1,1)+np.random.lognormal(-1,1)+np.random.lognormal(-1,1)+np.random.lognormal(-1,1)+np.random.lognormal(-1,1)+np.random.lognormal(-1,1)+np.random.lognormal(-1,1)+np.random.lognormal(-1,1)+np.random.lognormal(-1,1)+np.random.lognormal(-1,1))
       
plt.hist(c, normed=True)

d=np.mean(c)
e=np.var(c)
h = np.linspace(d-2*e, d+3*e, 100)
plt.plot(h, norm.pdf(h, loc=d, scale=np.sqrt(e)), c='red', lw=2)

plt.show()

```
