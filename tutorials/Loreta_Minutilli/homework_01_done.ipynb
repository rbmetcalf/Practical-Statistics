```python
# homework # 01

## Here we are going to get a little experience with the covariance matrix 
## of data and principle components analysis.  This was introduced in the 
## lectures.

# 1) Read file homework_01_2d-datafile.csv into a dataframe using pandas
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

filename = 'C:\\Users\\utente\\Desktop\\files\\homework_01_2d-datafile.csv'
data = pd.read_csv(filename)
data.head()
x = data[['X']]
y = data[['Y']]

    # plot them as points
plt.scatter(x,y)
plt.xlabel('x variable')
plt.ylabel('y variable')
plt.show()

# 2) Find the covariance matrix for the two variables.  There is a python function for this, 
#    but it would be more instructive if you write your own.
point_number = len(x)
x_np = np.array(x)
y_np = np.array(y)
variables = [x_np,y_np]
cov_matrix = np.empty([2,2])

for i in range(2) : 
    for j in range(2) : 
        cov_matrix[i][j] = np.sum((variables[i]-np.mean(variables[i]))*(variables[j]-np.mean(variables[j])))/point_number

print("Covariance matrix: " + str(cov_matrix))
     
    # Are they correlated? (Remember to normalize the correlation coefficient.)
corr_coeff = cov_matrix[0][1]/(np.std(x_np)*np.std(y_np))
print("Pearson coefficient: " + str(corr_coeff) + " The two variables are correlated.") 

# 3) Decompose the covariance matrix using an eigenvalue decompositions (There are python functions for this.)
cov_eig = np.linalg.eig(cov_matrix)
print("Eigenvalues: " + str(cov_eig[0]))
print("Eigenvectors: " + str(cov_eig[1]))
    # What are the principle components of the data?
print("The principal component of the data is the eigenvector corresponding to the highest eigenvalue.")    
    # What are the variances of each principle component?
print("The variance of each principle component are the corresponding eigenvalues.")

# 4) Do 1 through 3, but using the data file homework_01_5d-datafile.csv this time.
#    In this case the data is 5 dimensional.
filename5 = 'C:\\Users\\utente\\Desktop\\files\\homework_01_5d-datafile.csv'
data5 = pd.read_csv(filename5)
x1 = data5[['X1']]
x2 = data5[['X2']]
x3 = data5[['X3']] 
x4 = data5[['X5']]
x5 = data5[['X5']]

variables5 = [np.array(x1),np.array(x2),np.array(x3),np.array(x4),np.array(x5)]
point_number5 = len(x1)
cov_matrix5 = np.empty([5,5])

for i in range(5) :
    for j in range(4) : 
        if i!=j : 
            plt.scatter(variables5[i],variables5[j])
            plt.title(str(i+1) + ',' + str(j+1))
            plt.show()

for i in range(5) : 
    for j in range(5) : 
        cov_matrix5[i][j] = np.sum((variables5[i]-np.mean(variables5[i]))*(variables5[j]-np.mean(variables5[j])))/point_number5

print("Covariance matrix: " + str(cov_matrix5))


for i in range(5) : 
    for j in range(4) : 
        if i!=j : 
            corr_coeff5 = cov_matrix5[i][j]/(np.std(variables5[i])*np.std(variables5[j]))
            if corr_coeff5 > 0 :
                print("The variables " + str(i+1) + " and " + str(j+1) + " are correlated. The correlation coefficient is " 
                      + str(corr_coeff5))

cov_eig5 = np.linalg.eig(cov_matrix5)
print("Eigenvalues: " + str(cov_eig5[0]))
print("Eigenvectors: " + str(cov_eig5[1]))

    # Which variables seems to be correlated with each other and which ones not?
```
