```python
#Tutorial number 1:-
#-- Create a scatter plot of a gaussian data (hint: see numpy funtion random.normal) of size 1000,
#-- located at 0.0 and with scale 5.0.

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import NullFormatter
%matplotlib inline

x_gauss = np.random.normal(0.0, 5.0, 1000)
y_gauss = np.random.normal(0.0, 5.0, 1000)

#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~
#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~

#-- Enhance the plot with axis labels  (x-position , y-position) and don't forget the units in Mpc/h, 
#-- plot title ("2d galaxy distribution") and x&y-axis ticks [-5, -3, .., 3, 5] and set dimond as point marker, 
#-- blue as point color and  1.5 for  point size. 

plt.figure(1, figsize=(7, 7))

plt.xlabel('x-position  [Mpc/h]')
plt.ylabel('y-position  [Mpc/h]')
plt.title('2d galaxy distribution')
plt.xlim(-20, 20)
plt.ylim(-20, 20)
ticks = np.linspace(-20, 20, 9)
plt.xticks(ticks, ticks)
plt.yticks(ticks, ticks)

plt.scatter(x_gauss, y_gauss, s=1.5, c='b', marker='D')

plt.show()

#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~
#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~

#-- Create a histogram of the previous x&y data with 20 equispaced linear bins and plot it.

plt.figure(1, figsize=(7, 4))

xbins = np.arange(x_gauss.min(), x_gauss.max(), (x_gauss.max()-x_gauss.min())/20.)
plt.hist(x_gauss, xbins, color='green')

plt.xlabel('x-position  [Mpc/h]')
plt.ylabel('Number')
plt.title('x-position galaxy distribution')
plt.xticks(ticks, ticks)

plt.show()

#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~

plt.figure(1, figsize=(7, 4))

ybins = np.arange(y_gauss.min(), y_gauss.max(), (y_gauss.max()-y_gauss.min())/20.)
plt.hist(y_gauss, ybins, color='green')

plt.xlabel('y-position  [Mpc/h]')
plt.ylabel('Number')
plt.title('y-position galaxy distribution')
plt.xlim(-20, 20)
plt.xticks(ticks, ticks)

plt.show()

#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~
#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~

#-- Creat a combined plot that shows the scatter data in the middel with x&y historgrams beside each axis. 
#-- Look at the follwoing example for help (https://matplotlib.org/examples/pylab_examples/scatter_hist.html)

plt.figure(1, figsize=(10, 10))

left, width = 0.1, 0.65
bottom, height = 0.1, 0.65
bottom_h = left_h = left + width + 0.02

rect_scatter = [left, bottom, width, height]
rect_histx = [left, bottom_h, width, 0.2]
rect_histy = [left_h, bottom, 0.2, height]

axScatter = plt.axes(rect_scatter)
axHistx = plt.axes(rect_histx)
axHisty = plt.axes(rect_histy)

nullfmt = NullFormatter()

axHistx.xaxis.set_major_formatter(nullfmt)
axHisty.yaxis.set_major_formatter(nullfmt)

axScatter.scatter(x_gauss, y_gauss, s=1.5, c='b', marker='D')
axScatter.set_xlim(-20, 20)
axScatter.set_ylim(-20, 20)
axScatter.set_xlabel('x-position  [Mpc/h]')
axScatter.set_ylabel('y-position  [Mpc/h]')

axHistx.hist(x_gauss, bins=20, color='b')
axHistx.set_ylabel('Number')

axHisty.hist(y_gauss, bins=20, color='b', orientation='horizontal')
axHisty.set_xlabel('Number')

plt.show()
```
