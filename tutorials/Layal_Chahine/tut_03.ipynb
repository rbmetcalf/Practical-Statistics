```python
#Tutorial # 03

## The central limit theorem holds that the sum of a large number of identically 
## distributed random numbers will be normally distributed no matter how the 
## individual numbers are distributed.
##
## In lecture we proved this and saw how the pdf for the sum of variables 
## can be found numerically from the pdf of the individuals using the Fourier 
## transform.
##
## Now we will verify the central limit theorem by numerically experiment :

# 1) Create 1000 uniform random numbers between -1 and 1 using numpy.rand
import numpy as np
a=np.random.uniform(-1,1,1000)
       # make histogram of the numbers
import matplotlib.pyplot as plt
_=plt.hist(a,normed=True)
plt.title('uniform random numbers')
plt.show()
# 2) Make 1000 sums of 3 uniform random numbers between -1 and 1
sum1=np.zeros(1000)
for i in range(2) :
     sum1 = a+np.random.uniform(-1,1,1000)

       # make histogram of the numbers
plt.hist(sum1,normed=True,bins=10)
plt.title('sum on 3 uniform random numbers')

       # draw the pdf for a normal distribution with the same man and variance
        # over histogram (What should the variance be?)
var=np.var(sum1)
mean=np.mean(sum1)
std=np.sqrt(var)
pdf=np.random.normal(mean,std,size=1000)
plt.hist(pdf,bins=100,normed=True,histtype='step')
plt.show()
# 3) Make 1000 sums of 10 uniform random numbers between -1 and 1
sum2=np.zeros(1000)
for i in range(9) :
    sum2 = a + np.random.uniform(-1,1,1000)
       # make histogram of the numbers
plt.hist(sum2,normed=True,bins=10)
plt.title('sum on 10 uniform random numbers')
       # draw the pdf for a normal distribution with the same variance over histogram
var=np.var(sum2)
std=np.sqrt(var)
mean=np.mean(sum2)
pdf=np.random.normal(mean,std,size=1000)
plt.hist(pdf,bins=100,normed=True,histtype='step')
plt.show()
# 4) Do 1 through 3 for a lognormal distribution instead of uniform.  The 
#    lognormal distribution can be found in the scipy.stats package. Use 
#    the case s = 1.  An normalize the sum with 1/N where N is the number of 
#    numbers in each sum.  Draw the Gaussian with the same mean and variance as 
#    the sample.
# 1)
from scipy.stats import lognorm
b = lognorm.rvs(1, size=1000)
plt.hist(b,normed=True)
plt.title('lognormal random numbers')
plt.show()
# 2)
sum_1=np.zeros(1000)
for i in range(2) :
    sum_1 = b + lognorm.rvs(1, size=1000)
plt.hist(sum_1,normed=True,bins=10)
plt.title('sum on 3 lognormal random numbers ')
var=np.var(sum_1)
mean=np.mean(sum_1)
std=np.sqrt(var)
pdf=np.random.normal(mean,std,size=1000)
plt.hist(pdf,bins=100,normed=True,histtype='step')
plt.show()
# 3)
sum_2=np.zeros(1000)
for i in range(9) :
    sum_2 = b + lognorm.rvs(1, size=1000)
plt.hist(sum_2,normed=True,bins=10)
plt.title('sum on 10 lognormal random numbers ')
var=np.var(sum_2)
mean=np.mean(sum_2)
std=np.sqrt(var)
pdf=np.random.normal(mean,std,size=1000)
plt.hist(pdf,bins=100,normed=True,histtype='step')
plt.show()
```
