```python
#Tutorial # 03

## The central limit theorem holds that the sum of a large number of identically 
## distributed random numbers will be normally distributed no matter how the 
## individual numbers are distributed.
##
## In lecture we proved this and saw how the pdf for the sum of variables 
## can be found numerically from the pdf of the individuals using the Fourier 
## transform.
##
## Now we will verify the central limit theorem by numerically experiment :
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
from scipy.stats import lognorm
% matplotlib inline

# 1) Create 1000 uniform random numbers between -1 and 1 using numpy.rand
x_0=np.random.uniform(-1.,1.,1000)
# make histogram of numbers
plt.hist(x_0,bins=int(np.sqrt(len(x_0))),normed=True)
plt.title('One single uniform distribution')
plt.show()
        

# 2) Make 1000 sums of 3 uniform random numbers between -1 and 1
x_1=np.random.uniform(-1.,1.,1000)
x_2=np.random.uniform(-1.,1.,1000)
x_3=np.random.uniform(-1.,1.,1000)
sum_0=x_1+x_2+x_3
# make histogram of numbers
plt.hist(sum_0,bins=int(np.sqrt(len(sum_0))),normed=True,label='Sum of three uniform distributons')
# draw normal distribution with the same mean and variance over histogram 
# (What should the variance be?)
mean_0=np.mean(sum_0)
var_0=np.var(sum_0)
x_0=np.linspace(mean_0-5*var_0,mean_0+5*var_0,300)
gauss_0=plt.plot(x_0,mlab.normpdf(x_0,mean_0,np.sqrt(var_0)),label='Normal distribution')
plt.title('Comparing the sum of three uniform distributions with a single normal distributions')
plt.legend(loc='lower center')
plt.show()


# 3) Make 1000 sums of 10 uniform random numbers between -1 and 1
x_4=np.random.uniform(-1.,1.,1000)
x_5=np.random.uniform(-1.,1.,1000)
x_6=np.random.uniform(-1.,1.,1000)
x_7=np.random.uniform(-1.,1.,1000)
x_8=np.random.uniform(-1.,1.,1000)
x_9=np.random.uniform(-1.,1.,1000)
x_10=np.random.uniform(-1.,1.,1000)
x_11=np.random.uniform(-1.,1.,1000)
x_12=np.random.uniform(-1.,1.,1000)
x_13=np.random.uniform(-1.,1.,1000)
sum_1=x_4+x_5+x_6+x_7+x_8+x_9+x_10+x_11+x_12+x_13
       # make histogram of numbers
plt.hist(sum_1,bins=int(np.sqrt(len(sum_1))),normed=True,label='Sum of ten uniform distributions')
       # draw normal distribution with the same variance over histogram
mean_1=np.mean(sum_1)
var_1=np.var(sum_1)
x_1=np.linspace(mean_1-5*var_1,mean_1+5*var_1,300)
gauss_1=plt.plot(x_1,mlab.normpdf(x_1,mean_1,np.sqrt(var_1)),label='Normal distribution')
plt.title('Comparing the sum of ten uniform distributions with a single normal distribution')
plt.legend(loc='lower center')
plt.show()


# 4) Do 1 through 3 for a lognormal distribution instead of uniform.  The 
#    lognormal distribution can be found in the scipy.stats package. 


#lognormal (1)
x_l0=np.random.lognormal(0.,1.,1000)
plt.hist(x_l0,bins=int(np.sqrt(len(x_l0))),normed=True)
plt.title('One single lognormal distribution')
plt.show()


#lognormal (2)
x_l1=np.random.lognormal(0.,1.,1000)
x_l2=np.random.lognormal(0.,1.,1000)
x_l3=np.random.lognormal(0.,1.,1000)
sum_l0=x_l1+x_l2+x_l3
plt.hist(sum_l0,bins=int(np.sqrt(len(sum_l0))),normed=True,label='Sum of three lognormal distributons')
mean_l0=np.mean(sum_l0)
var_l0=np.var(sum_l0)
x_l0=np.linspace(mean_l0-var_l0,mean_l0+2*var_l0,300)
logn_0=plt.plot(x_l0,mlab.normpdf(x_l0,mean_l0,np.sqrt(var_l0)),label='Normal distribution')
plt.title('Comparing the sum of three lognormal distributions with a single normal distribution')
plt.legend()
plt.show()


#lognormal (3)
x_l4=np.random.lognormal(0.,1.,1000)
x_l5=np.random.lognormal(0.,1.,1000)
x_l6=np.random.lognormal(0.,1.,1000)
x_l7=np.random.lognormal(0.,1.,1000)
x_l8=np.random.lognormal(0.,1.,1000)
x_l9=np.random.lognormal(0.,1.,1000)
x_l10=np.random.lognormal(0.,1.,1000)
x_l11=np.random.lognormal(0.,1.,1000)
x_l12=np.random.lognormal(0.,1.,1000)
x_l13=np.random.lognormal(0.,1.,1000)
sum_l1=x_l4+x_l5+x_l6+x_l7+x_l8+x_l9+x_l10+x_l11+x_l12+x_l13
plt.hist(sum_l1,bins=int(np.sqrt(len(sum_l1))),normed=True,label='Sum of ten lognormal distributions')
mean_l1=np.mean(sum_l1)
var_l1=np.var(sum_l1)
x_l1=np.linspace(mean_l1-0.5*var_l1,mean_l1+1*var_l1,300)
logn_1=plt.plot(x_l1,mlab.normpdf(x_l1,mean_l1,np.sqrt(var_l1)),label='Normal distribution')
plt.title('Comparing the sum of ten lognormal distributions with a single normal distribution')
plt.legend()
plt.show()
```
